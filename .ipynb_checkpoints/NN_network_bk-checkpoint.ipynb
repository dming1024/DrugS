# 搭建模型

```python
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow as tf
import pandas as pd
from tensorflow.keras.layers import Concatenate, LSTM, GRU
#call back
from tensorflow.keras import layers, callbacks
#token化处理
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.models import Sequential

import re
import numpy as np
import gc

from tensorflow import keras
from tensorflow.keras import layers
import tensorflow as tf
import pandas as pd
from tensorflow.keras.optimizers import Adam

#call back
from tensorflow.keras import layers, callbacks
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import ReduceLROnPlateau

#使用keras_tuner进行参数调整
from keras_tuner.tuners import Hyperband
from sklearn.model_selection import train_test_split
```

```python
#数据准备
padded_sequences=np.load('padded_sequences.npy')
padded_sequences=np.array(padded_sequences).reshape(-1,32,30,20)
compount_fp_array=np.load('compount_fp_array.npy')
IC50_res_array=np.load('IC50_res_array.npy')
```

```python
#搭建model
def build_model(hp):

    gene_input=keras.Input(shape=(32,30,20), name="gene")
    gene_features=layers.Conv2D(64, (3, 3),name='filter_x', activation='elu',kernel_regularizer=l2(0.01))(gene_input)
    gene_features=layers.BatchNormalization()(gene_features)
    gene_features=layers.Dropout(0.2)(gene_features)
    gene_features=layers.MaxPool2D((2,2), padding='same')(gene_features)
    
    for i in range(hp.Int('conv_layers', 1,3)):
        gene_features=layers.Conv2D(filters=hp.Int('filters_' + str(i),64,256,step=32),
                                     kernel_size=(3,3),
                                     activation='relu')(gene_features)
        gene_features=layers.BatchNormalization()(gene_features)
        gene_features=layers.Dropout(rate=hp.Float('dropout_' + str(i),
                                        min_value=0.1,
                                        max_value=0.5,
                                        step=0.1))(gene_features)
        gene_features=layers.MaxPool2D((2,2), padding='same')(gene_features)   
    
    gene_features=layers.GlobalAveragePooling2D()(gene_features)
    gene_features=layers.Dense(16, activation='elu')(gene_features)
    compound_input = keras.Input(shape=(881), name="compound") 
    compound_layers = layers.Dense(512, name="compound_0",kernel_regularizer=l2(0.01), activation='elu')(compound_input)
    compound_layers = layers.Dense(256, name="compound_1",kernel_regularizer=l2(0.01), activation='elu')(compound_layers)
    compound_layers = layers.Dense(16, name="compound_3", activation='elu')(compound_layers)


    x = layers.concatenate([gene_features, compound_layers])
    outputs = layers.Dense(1, name="ic50")(x)
    model = keras.Model(
        inputs=[gene_input,compound_input],
        outputs=[outputs],
    )
    
    learning_rate = hp.Choice('learning_rate', [1e-3, 1e-4, 1e-5])
    
    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss=keras.losses.MeanSquaredError(),
        metrics=keras.metrics.MeanAbsoluteError()
    )
    return model
    
tuner = Hyperband(
    build_model,
    objective='val_loss',
    max_epochs=30,
    factor=3,
    directory='my_dir',
    project_name='hyperband_tuning_A'
)
```

```python
# 获取最佳超参数
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

# 使用最佳超参数构建模型
model = tuner.hypermodel.build(best_hps)
```

```python
# 训练模型
early_stopping = callbacks.EarlyStopping(
    min_delta=0.0001, # minimium amount of change to count as an improvement
    patience=5, # how many epochs to wait before stopping
    restore_best_weights=True,
)
history=model.fit(
    {"gene": padded_sequences, "compound": compount_fp_array},
    {'ic50':IC50_res_array},
    epochs=10,
    batch_size=64,
    validation_split=0.3,
    callbacks=[early_stopping]
)
```

## 基于PCA 的 DNN网络分析

```python
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow as tf
import pandas as pd
from tensorflow.keras.optimizers import Adam

#call back
from tensorflow.keras import layers, callbacks
from tensorflow.keras.regularizers import l2

from tensorflow.keras import layers, models
```

```python
gene_input=keras.Input(shape=(1000,), name="gene")
gene_features=layers.Dense(512,kernel_regularizer=l2(0.01), activation="elu")(gene_input)
gene_features=layers.Dense(256,kernel_regularizer=l2(0.01), activation="elu")(gene_features)
gene_features=layers.Dense(128,kernel_regularizer=l2(0.01), activation="elu")(gene_features)
gene_features=layers.Dense(64,kernel_regularizer=l2(0.01), activation="elu")(gene_features)
gene_features=layers.Dense(16,kernel_regularizer=l2(0.01), activation="elu")(gene_features)
```

```python
compound_input = keras.Input(shape=(881), name="compound") 
compound_layers = layers.Dense(512, name="compound_0",kernel_regularizer=l2(0.01), activation="elu")(compound_input)
compound_layers = layers.Dense(256, name="compound_1",kernel_regularizer=l2(0.01), activation="elu")(compound_layers)
compound_layers = layers.Dense(128, name="compound_2",kernel_regularizer=l2(0.01), activation="elu")(compound_layers)
compound_layers = layers.Dense(16, name="compound_3", activation="elu")(compound_layers)
```

```python
x = layers.concatenate([gene_features, compound_layers])
outputs = layers.Dense(1, name="ic50")(x)
model = keras.Model(
    inputs=[gene_input,compound_input],
    outputs=[outputs],
)
model.compile(
    #optimizer=keras.optimizers.RMSprop(1e-3),
    optimizer=Adam(learning_rate=0.001),
    loss=keras.losses.MeanSquaredError(),
    metrics=keras.metrics.MeanAbsoluteError()
)
early_stopping = callbacks.EarlyStopping(
    min_delta=0.001, # minimium amount of change to count as an improvement
    patience=20, # how many epochs to wait before stopping
    restore_best_weights=True,
)
model.summary()
```

### 使用Autoencoder进行基因表达降维处理

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
```

```python
padded_sequences=np.load("padded_sequences_zscore.npy")
```

```python
padded_sequences.shape
```

```python
gene_expression=pd.DataFrame(padded_sequences.reshape(-1,19200))
```

```python
input_dim = gene_expression.shape[1]  # 输入数据的特征维度
encoding_dim = 64  # 压缩到的维度

# 输入层
input_layer = Input(shape=(input_dim,))

# 编码层
encoded = Dense(2048, activation='elu')(input_layer)
encoded = Dense(512, activation='elu')(encoded)
encoded = Dense(64, activation='elu')(encoded)
encoded = Dense(encoding_dim, activation='elu')(encoded)

# 解码层
decoded = Dense(64, activation='elu')(encoded)
decoded = Dense(512, activation='elu')(decoded)
decoded = Dense(2048, activation='elu')(decoded)
decoded = Dense(input_dim, activation='sigmoid')(decoded)

# 自动编码器模型
autoencoder = Model(input_layer, decoded)

# 编码器模型，用于降维
encoder = Model(input_layer, encoded)

# 编译模型
autoencoder.compile(optimizer='adam', loss='mse')

early_stopping = callbacks.EarlyStopping(
    min_delta=0.001, # minimium amount of change to count as an improvement
    patience=5, # how many epochs to wait before stopping
    restore_best_weights=True,
)
```

```python
# 训练模型
autoencoder.fit(gene_expression, gene_expression, epochs=50, batch_size=256, shuffle=True, validation_split=0.2, callbacks=[early_stopping],)
```

```python
#进行scale转换
from scipy.stats import zscore
gene_expression=expression_CCLE.drop(['ccl_name'],axis=1).apply(zscore,axis=0)#1是逐行，默认是逐列(0)
```

```python
#standard scaler转换
#from sklearn.preprocessing import StandardScaler
#scaler = StandardScaler()
#gene_expression = scaler.fit_transform(expression_CCLE.drop(['ccl_name'],axis=1))
```

```python
input_dim = gene_expression.shape[1]  # 输入数据的特征维度
encoding_dim = 30  # 压缩到的维度
```

```python
from tensorflow.keras.regularizers import l2

# 设置随机种子
tf.random.set_seed(42)
np.random.seed(42)

# 输入层
input_layer = keras.Input(shape=(input_dim,))
# 编码层
encoded = Dense(2048, activation='elu')(input_layer)
encoded = Dense(encoding_dim, activation='relu')(encoded)
# 解码层
decoded = Dense(2048, activation='elu')(encoded)
decoded = Dense(input_dim, activation='relu')(decoded)
# 自动编码器模型
autoencoder = Model(input_layer, decoded)
# 编码器模型，用于降维
encoder = Model(input_layer, encoded)
# 编译模型
autoencoder.compile(optimizer='adam', loss='mse')
early_stopping = callbacks.EarlyStopping(
    min_delta=0.001, # minimium amount of change to count as an improvement
    patience=10, # how many epochs to wait before stopping
    restore_best_weights=True,
)
# 训练模型
autoencoder.fit(gene_expression, gene_expression, epochs=50, 
                batch_size=256, shuffle=True, validation_split=0.2, callbacks=[early_stopping],)
```

### 数据合并

```python
import re
def merge_data(GSVA_Score,modelID,fingerprint,IC50):
    #step1: merge score with modelID
    GSVA_Score_modelID=GSVA_Score.merge(
    modelID,
    how='inner',
    left_on='ccl_name', #确认名称是否正确
    right_on='ModelID' 
    )
    #step2: merge ic50 with fingerprint
    ic50_fingerprint=IC50.merge(
    fingerprint,
    how='inner',
    left_on='DRUG_ID',#确认名称是否正确
    right_on='Name'
    )
    #step3: 根据细胞系名称合并 gsva数据和ic50数据
    ic50_fingerprint['stripped_name']=[re.sub('\\-','',m) for m in ic50_fingerprint.CELL_LINE_NAME]
    data=ic50_fingerprint.merge(
        GSVA_Score_modelID,
        how='inner',
        left_on='stripped_name',
        right_on= 'StrippedCellLineName'
    )
    return data
```

```python
def build_model(hp):
    model = Sequential()
    
    # 输入层
    model.add(Dense(X_train.shape[1], input_dim=X_train.shape[1], activation='relu'))
    
    # 第一隐藏层，固定大小为512
    model.add(Dense(512, activation='relu'))
    
    # 可变隐藏层和神经元数量
    for i in range(hp.Int('num_layers', 2, 6)):
        model.add(Dense(units=hp.Int('units_' + str(i),
                                     min_value=32,
                                     max_value=256,
                                     step=32),
                        activation='relu'))
        model.add(Dropout(rate=hp.Float('dropout_' + str(i),
                                        min_value=0.1,
                                        max_value=0.5,
                                        step=0.1)))
    
    # 输出层
    model.add(Dense(1))
    
    # 优化器
    learning_rate = hp.Choice('learning_rate', [1e-3,1e-5])
    model.compile(optimizer=Adam(learning_rate=learning_rate),
                  loss=MeanSquaredError(),
                  metrics=['mse'])
    
    return model
```
